### Problem

Using a suitable Python library, extrapolate the time series of the Paladini example over the next 6 months.

### Resolution

We can extrapolate the decomposition through the python library _statsmodels_. In particoular we can use the function _seasonal_decompose()_ in order to extract and print the 4 variables.

```python
from random import randrange
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
%matplotlib inline
```

```python
file = "Paladini  Associates example.csv"

data = open(file, "r")
data = data.read()
data = data.split(",")
data = [float(d) for d in data]
data = np.array(data)
```

```python
m = data.shape[0]
```

```python
pal = seasonal_decompose(data, model='multiplicative', freq=12)
pal.plot()
plt.show()
```

```python
M = 12
T = data.shape[0]
qvt = np.zeros(data.shape[0])
for i in range(int(M/2),T-int(M/2)):
    tmp = 0
    tmp = data[i-6]/2
    tmp += data[i+6]/2
    for j in range(i-5,i+6):
        tmp += data[j]
    qvt[i] = tmp/M
```

```python
x = np.arange(qvt.shape[0])
plt.figure(figsize=(15,7))
plt.plot(x[int(M/2):T-int(M/2)], qvt[int(M/2):T-int(M/2)])
plt.show()
```

```python
z = np.polyfit(x[int(M/2):T-int(M/2)], qvt[int(M/2):T-int(M/2)], 1)
```

```python
q = np.zeros(data.shape[0])
q[int(M/2):T-int(M/2)] = z[0] * x[int(M/2):T-int(M/2)] + z[1]
```

```python
v = np.zeros(data.shape[0])
v[int(M/2):T-int(M/2)] = qvt[int(M/2):T-int(M/2)] / q[int(M/2):T-int(M/2)]
```

```python
srt = np.zeros(data.shape[0])
srt[int(M/2):T-int(M/2)] = data[int(M/2):T-int(M/2)]/qvt[int(M/2):T-int(M/2)]
```

```python
s = np.zeros(12)
for i in range(12):
    count = 0
    for j in range(i+12,T-int(M/2)-1,12):
        s[i] += srt[j]
        count += 1
    s[i] = s[i]/count
```

```python
xx = v[149:157]
yy=np.arange(xx.shape[0])
zz = np.polyfit(yy, xx, 2)
```

```python
t = np.arange(xx.shape[0],xx.shape[0]+7)
a = zz[0] * (t**2) + zz[1]*t + zz[2]
```

```python
t = np.arange(157,163)
qt = z[0]*t+z[1]
```
