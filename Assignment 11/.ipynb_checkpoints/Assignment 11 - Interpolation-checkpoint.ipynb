### Problem

Using a proper Python library, develop an interpolation model for the example 1 presented in the regression lecture.
Plot and compare the results.

### Resolution

We can use a function from the _scipy_ library called _lagrange()_ that implement automatically the Lagrange interpolation.

```python
import matplotlib.pyplot as plt
from scipy.interpolate import lagrange

x = [10.11, 50.56, 90.28, 15.50, 69.52, 98.40, 86.66, 42.87, 61.53, 24.60, 46.85, 50.63, 89.68, 27.91]
y = [1.53, 13.14, 31.24, 5.47, 22.27, 26.47, 24.32, 13.51, 23.65, 9.43, 15.12, 18.94, 26.06, 10.08]

poly = lagrange(x, y)
print(poly)
```

```python
xnew=[]
for i in range (10,100):
    xnew.append(i)
plt.figure(figsize=(15, 10))
plt.plot(x,y, 'ro')
plt.plot(x,y, 'b')
plt.plot(xnew, poly(xnew), color='orange')
plt.title('Interpolation of Lagrange')
plt.xlabel('Volume')
plt.ylabel('Cost')

plt.show()
```
